```python
import pandas as pd
import numpy as np
import os
import glob
pd.set_option('display.max_columns', 10)

# in the folder
path = os.getcwd()
csv_files = glob.glob(os.path.join("../*simulation00*"))
```

```python
%%time
dfs = dict()
for f in csv_files:
    # read the csv file
    try:
        df = pd.read_csv(f,delimiter=' ',index_col=False)
        dfs[f]=df
    except:
        print('Reading error for '+f)
```

```python
df_V = dict()
for f in dfs:
    if dfs[f].shape[0]==1001:
        df_V[f] = dfs[f].V.to_list()
df_V = pd.DataFrame(df_V)
df_V = df_V.T
```

```python
df_V.to_csv('df_V_sim00.csv')
```

```python
import matplotlib.pyplot as plt
for f in df_V.index:
    try:
        plt.plot(dfs[f]['Time'],dfs[f]['V'])
    except:
        print('read error',f)
```
